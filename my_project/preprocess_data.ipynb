```python
def merge_data(experiment, time_window_size = 0.03):
    df_acc = pd.read_csv(f'data\{experiment}\Accelerometer.csv')
    df_acc['Time window'] = np.floor(df_acc['Time (s)'] / time_window_size) * time_window_size

    df_gyro = pd.read_csv(f'data\{experiment}\Gyroscope.csv')
    df_gyro['Time window'] = np.floor(df_gyro['Time (s)'] / time_window_size) * time_window_size

    df_mag = pd.read_csv(f'data\{experiment}\Magnetometer.csv')
    df_mag['Time window'] = np.floor(df_mag['Time (s)'] / time_window_size) * time_window_size

    df_lin_acc = pd.read_csv(f'data\{experiment}\Linear Acceleration.csv')
    df_lin_acc['Time window'] = np.floor(df_lin_acc['Time (s)'] / time_window_size) * time_window_size

    df_prox = pd.read_csv(f'data\{experiment}\Proximity.csv')
    df_prox['Time window'] = np.floor(df_prox['Time (s)'] / time_window_size) * time_window_size

    df_full = pd.merge(df_acc, df_gyro, how='outer', on='Time window', suffixes=('_acc', '_gyro'))
    df_full = pd.merge(df_full, df_mag, how='outer', on='Time window', suffixes=('', '_mag'))
    df_full = pd.merge(df_full, df_lin_acc, how='outer', on='Time window', suffixes=('', '_lin_acc'))
    df_full = pd.merge(df_full, df_prox, how='outer', on='Time window', suffixes=('', '_prox'))

    df_full.sort_values(inplace=True, by='Time window')
    df_full.reset_index(inplace=True)

    df_full['Time (s)_mag'] = df_full['Time (s)']
    try:
        df_full = df_full[['Time window', 'Time (s)_acc', 'Time (s)_gyro', 'Time (s)_mag', 'Time (s)_lin_acc', 'Time (s)_prox', 
                    'Acceleration x (m/s^2)', 'Acceleration y (m/s^2)', 'Acceleration z (m/s^2)', 
                    'Gyroscope x (rad/s)', 'Gyroscope y (rad/s)', 'Gyroscope z (rad/s)',
                    'Magnetic field x (µT)', 'Magnetic field y (µT)', 'Magnetic field z (µT)',
                    'Linear Acceleration x (m/s^2)', 'Linear Acceleration y (m/s^2)', 'Linear Acceleration z (m/s^2)',
                    'Distance (cm)']]
    except KeyError:
        df_full = df_full[['Time window', 'Time (s)_acc', 'Time (s)_gyro', 'Time (s)_mag', 'Time (s)_lin_acc', 'Time (s)_prox', 
                    'X (m/s^2)', 'Y (m/s^2)', 'Z (m/s^2)', 
                    'X (rad/s)', 'Y (rad/s)', 'Z (rad/s)',
                    'X (µT)', 'Y (µT)', 'Z (µT)',
                    'X (m/s^2)_lin_acc', 'Y (m/s^2)_lin_acc', 'Z (m/s^2)_lin_acc',
                    'Distance (cm)']]
        df_full.rename(columns={k:v for k, v in zip(
            ['X (m/s^2)', 'Y (m/s^2)', 'Z (m/s^2)', 
            'X (rad/s)', 'Y (rad/s)', 'Z (rad/s)',
            'X (µT)', 'Y (µT)', 'Z (µT)',
            'X (m/s^2)_lin_acc', 'Y (m/s^2)_lin_acc', 'Z (m/s^2)_lin_acc'],
            ['Acceleration x (m/s^2)', 'Acceleration y (m/s^2)', 'Acceleration z (m/s^2)', 
            'Gyroscope x (rad/s)', 'Gyroscope y (rad/s)', 'Gyroscope z (rad/s)',
            'Magnetic field x (µT)', 'Magnetic field y (µT)', 'Magnetic field z (µT)',
            'Linear Acceleration x (m/s^2)', 'Linear Acceleration y (m/s^2)', 'Linear Acceleration z (m/s^2)']
        )}, inplace=True)
    return df_full
    
```

```python
experiments = [
    'cycling 2024-06-07 12-40-37',
    'cycling 2024-06-07 12-45-31', 
    'cycling 2024-06-08 12-45-52',
    'running 2024-06-09 12-44-51', 
    'running 2024-06-09 12-49-21', 
    'walking 2024-06-09 11-28-48', 
    'walking 2024-06-09 11-42-12'
    ]

for exp in experiments:
    df = merge_data(exp)
    df.to_csv(f'data\{exp}\merged.csv', index=False)
```
